Two basic questions over a country population?

```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
import pandas as pd
```

```python
from scipy import stats
```

## Dataset from Kaggle ##

[Countries of the World](https://www.kaggle.com/fernandol/countries-of-the-world) Data compiled from US Government 1970-2017 Fernando Lasso
 â€¢  updated 9 months ago (Version 1)

```python
countries = pd.read_csv('countries of the world.csv')
```

```python
population = countries['Population']
```

```python
population = np.sort(population)
```

```python
plt.plot(population, '+')
plt.title('Population')
plt.xlabel('Country (One)')
plt.ylabel('People (One))')
plt.grid(True)
```

```python
plt.semilogy(population, '+')
plt.title('Population')
plt.xlabel('Country (one)')
plt.ylabel('log(People) (one)')
plt.grid(True)
```

```python
plt.plot(np.cumsum(population))
plt.title('Accumulated Population')
plt.xlabel('Country')
plt.ylabel('People')
plt.grid(True)
```

```python
log_population = np.log10(population)
```

```python
x = np.arange(0, 227)
y = log_population
```

```python
m, c = np.polyfit(x[100:170], y[100:170], 1)
print('Linear Regresion {0:2.3f} x + {1:2.3f}'.format(m,c))
```

```python
a1 = 10
a2 = 90
gradient, intercept, r_value, p_value, std_err = stats.linregress(x[a1:a2],y[a1:a2])
mn=np.min(x[a1:a2])-10
mx=np.max(x[a1:a2])+20
x1=np.linspace(mn,mx,500)
y1=gradient*x1+intercept
print('Linear Regresion {0:2.3f} x + {1:2.3f} ({2:2.3f})'.format(gradient, intercept, r_value))
```

```python
a1 = 90
a2 = 215
gradient, intercept, r_value, p_value, std_err = stats.linregress(x[a1:a2],y[a1:a2])
mn=np.min(x[a1:a2])-20
mx=np.max(x[a1:a2])
x2=np.linspace(mn,mx,500)
y2=gradient*x2+intercept
print('Linear Regresion {0:2.3f} x + {1:2.3f} ({2:2.3f})'.format(gradient, intercept, r_value))
```

```python
plt.plot(x,y,'+b')
plt.plot(x1,y1,'-r')
plt.plot(x2,y2,'-r')
plt.xlabel('Country (one)')
plt.ylabel('log10(People) (one)')
plt.grid(True)
plt.show()
```
